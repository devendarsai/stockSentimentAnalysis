```python
ticker_List = ["TSLA", "RBLX", "GFS", "UPST", "TITN"]
#File and directory handling to store output csv files
# path = "C:\\"
#Code to recursively remove the folder and its contents if it already exists
shutil.rmtree("./output", ignore_errors=False, onerror=None)
#Create a new directory in specified path to store the csv files
os.mkdir("./output")
```

```python
#Lines 37-45: Traverse through all the items of the list to extract the tweets and store in a csv file 
for i in range(len(ticker_List)):
    # Open/Create a file to append data
    csvFile = open("./output/" + ticker_List[i] + ".csv", 'w')
    #Use csv Writer
    csvWriter = csv.writer(csvFile)
    for tweet in tweepy.Cursor(api.search_tweets,q=("#" + ticker_List[i]),lang="en",since="2022-04-01" + "-is:RT").items():
        print (tweet.created_at, tweet.text)
        csvWriter.writerow([tweet.created_at, tweet.text])
    csvFile.close()
```
